```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#install c45 iblary
!pip install -U c45-decision-tree #https://pypi.org/project/c45-decision-tree/
```

```python
#two files are provided containing respecively train and test data sets
train_data = 'https://raw.githubusercontent.com/komisarzGiT/gai/refs/heads/main/GU_ML01/Assignment1/EEG_Train_Binary.csv'
test_data = 'https://raw.githubusercontent.com/komisarzGiT/gai/refs/heads/main/GU_ML01/Assignment1/EEG_Test_Binary.csv'

df_train = pd.read_csv(train_data)
df_test = pd.read_csv(test_data)

#check data types
print(f"Data types in training set \n{df_train.dtypes}")
print(f"\nData types in testing set \n{df_test.dtypes}")
```

```python
train_rows, train_columns = df_train.shape
print(f"Training data set has {train_rows} observations and {train_columns} variables. Sample data:")
display(df_train.head(3))

test_rows, test_columns = df_test.shape
print(f"\nTraining data set has {test_rows} observations and {test_columns} variables. Sample data:")
display(df_test.head(3))
```

```python
#check for missing values and outliers
print(f"\nSummary statistics to check for missing values and outliers in training data")
display(df_train.describe().round(2).T)

print(f"\nSummary sStatistics to check for missing values and outliers in test data")
display(df_test.describe().round(2).T)
```

```python
#split explanatory variables from target variables
X_train = df_train.drop('y', axis=1) #explanatory
y_train = df_train['y'] #target

X_test = df_test.drop('y', axis=1)
y_test = df_test['y']
```

```python
#count number of distinct values for all features
train_count_distinct = X_train.nunique()
print(f"Average number of distinct values across all features in training data set is: \n {train_count_distinct.mean()}")
print(f"\nCount of distinct values per feature \n{train_count_distinct}")

test_count_distinct = X_test.nunique()
print(f"\nAverage number of distinct values across all features in training data set is: \n {test_count_distinct.mean()}")
print(f"\nCount of distinct values per feature \n{test_count_distinct}")
```

```python
#Below code is based on lecture
#CT5148 Programming and Tools for Artificial Intelligence - Online
#by Dr James McDermott, University of Galway
#Week 5: scikit-learn

#Training and evaluating decision trees

from sklearn.tree import DecisionTreeClassifier
from C45 import C45Classifier

# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html
# criterion{“gini”, “entropy”, “log_loss”}, default=”gini”
# The function to measure the quality of a split. Supported criteria are “gini” for the Gini impurity and “log_loss” and “entropy” both for the Shannon information gain, see Mathematical formulation.

# Train DT Gini
dt_gini = DecisionTreeClassifier(criterion='gini', random_state=7)
dt_gini.fit(X_train, y_train)

# Evaluate DT Gini
accuracy_gini = dt_gini.score(X_test, y_test)
print(f"Accuracy DT Gini: {accuracy_gini:.4f}")

# Train DT Inf Gain
dt_inf_gain = DecisionTreeClassifier(criterion='entropy', random_state=7)
dt_inf_gain.fit(X_train, y_train)

# Evaluate DT Inf Gain
accuracy_inf_gain = dt_inf_gain.score(X_test, y_test)
print(f"Accuracy DT Information Gain: {accuracy_inf_gain:.4f}")

#Train DT Inf Gain Ratio
dt_inf_gain_ratio = C45Classifier()
dt_inf_gain_ratio.fit(X_train, y_train)

#Evaluate DT Inf Gain Ratio
y_pred_inf_gain_ratio = pd.Series(dt_inf_gain_ratio.predict(X_test))  #predict returned list, converting to series
total_matches = (y_test == y_pred_inf_gain_ratio).sum()
accuracy_inf_gain_ratio = total_matches / len(y_test)
print(f"Accuracy DT Information Gain Ratio: {accuracy_inf_gain_ratio:.4f}")

#dt_inf_gain_ratio.evaluate(X_test, y_test)

```

```python
#calculate confusion matrix for additional context
from sklearn.metrics import confusion_matrix

# CM for Gini
print("Confusion Matrix for DT Gini")
y_pred_gini = dt_gini.predict(X_test)
print(confusion_matrix(y_test, y_pred_gini))

#CM for Inf Gain
print("\nConfusion Matrix for DT Inf Gain")
y_pred_inf_gain = dt_inf_gain.predict(X_test)
print(confusion_matrix(y_test, y_pred_inf_gain))

# CM for Inf Gain Ratio
print("\nConfusion Matrix for DeT Inf Gain Ratio")
print(confusion_matrix(y_test, y_pred_inf_gain_ratio))
```

```python
#Below code is based on lecture
#CT5148 Programming and Tools for Artificial Intelligence - Online
#by Dr James McDermott, University of Galway
#Week 5: scikit-learn

# KNN algoritm needs standarization of values across dimesions based on training
# data set scikit-learn StandardScaler uses z-scaling to scale date
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# standardize training and test data
scaler = StandardScaler() #z-scaling using std dev and mean
X_train_std = scaler.fit_transform(X_train) #derive mean and std dev from X_train and apply
X_test_std = scaler.transform(X_test) #apply X_train mean and std dev standardization to X_test

#list of k values to test
k_values = [1, 3, 5, 9, 15, 21, 49, 101]

print("Training and evaluating KNN:")

accuracy_knn_max = 0
for k in k_values:
    # Train KNN model
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_std, y_train)

    # Evaluate
    y_pred_knn = knn.predict(X_test_std)
    total_matches_knn = (y_test == y_pred_knn).sum()
    accuracy_knn = (total_matches_knn / len(y_test))
    print(f"Accuracy with KNN (k={k}): {accuracy_knn:.4f}")
    if accuracy_knn_max < accuracy_knn:
      accuracy_knn_max = accuracy_knn
      k_best = k

print(f"\nBest KNN accuracy is {accuracy_knn_max:.4f} achieved for k={k_best} ")
```

```python
#table with result summary
summary_data = {
    ' ':      ['Decision Tree Gini                  ',
              'Decision Tree Information Gain      ',
              'Decision Tree Information Gain Ratio',
              'K-Nearest Neighbors (k=1)           '],
    'Accuracy': [accuracy_gini,
                 accuracy_inf_gain,
                 accuracy_inf_gain_ratio,
                 accuracy_knn_max] # KNN k=1
}

summary_df = pd.DataFrame(summary_data)
print(summary_df.to_string(index=False)) #string to remove df index
```
